```python
# from fwalker import fun
# from reader import readtxt
import os
import shutil                           # 
import random                           # 
import numpy as np                      # 
import matplotlib.pyplot as plt         # 
from nltk.corpus import stopwords       # 

cachedStopWords = stopwords.words("english")    # 

def fileWalker(path):
    # fileArray
    fileArray = []
    for root, dirs, files in os.walk(path):
        for fn in files:
            eachpath = str(root+'\\'+fn)
            fileArray.append(eachpath)
    return fileArray


def test_set_select():
    # spamham10test
    filepath = r'D:\email'
    testpath = r'D:\email\test'
    files = fileWalker(filepath)
    random.shuffle(files)
    top10 = files[:10]
    for ech in top10:
        ech_name = testpath+'\\'+('_'.join(ech.split('\\')[-2:]))  # _
        shutil.move(ech, testpath)  # echtestpath
        os.rename(testpath+'\\'+ech.split('\\')[-1], ech_name)  # echech_name,
        # print('%s moved' % ech_name)
    return


def test_set_clear():
    # testspamham
    filepath = r'D:\email'
    testpath = r'D:\email\test'
    files = fileWalker(testpath)
    for ech in files:
        ech_initial = filepath + '\\' + '\\'.join(' '.join(ech.split('\\')[-1:]).split('_'))  # 
        ech_move = filepath + '\\' + (' '.join(ech.split('\\')[-1:]).split('_'))[0]  # 
        shutil.move(ech, ech_move)  # echech_move
        os.rename(ech_move+'\\'+' '.join(ech.split('\\')[-1:]), ech_initial)  # 
        # print('%s moved' % ech)
    return


def readtxt(path, encoding):
    # encodingpathlines
    with open(path, 'r',  encoding='utf-8') as f:
        lines = f.readlines()
    return lines

def fileWalker(path):
    # pathfileArray
    fileArray = []
    for root, dirs, files in os.walk(path):
        for fn in files:
            eachpath = str(root+'\\'+fn)
            fileArray.append(eachpath)
    return fileArray

def email_parser(email_path):
    # clean_word
    punctuations = """,.<>()*&^%$#@!'";~`[]{}|\\/~+_-=?"""
    content_list = readtxt(email_path, 'gbk')
    content = (' '.join(content_list)).replace('\r\n', ' ').replace('\t', ' ')
    clean_word = []
    for punctuation in punctuations:
        content = (' '.join(content.split(punctuation))).replace('  ', ' ')
        clean_word = [word.lower()
                      for word in content.split(' ') if word.lower() not in cachedStopWords and len(word) > 2]
        # 
    return clean_word


def get_word(email_file):
    # email_fileappendword_listextendword_setset
    word_list = []
    word_set = []
    email_paths = fileWalker(email_file)
    for email_path in email_paths:
        clean_word = email_parser(email_path)
        word_list.append(clean_word)
        word_set.extend(clean_word)
        # print(set(word_set))
    return word_list, set(word_set)


def count_word_prob(email_list, union_set):
    # word_prob
    word_prob = {}
    for word in union_set:
        counter = 0
        for email in email_list:
            if word in email:
                counter += 1
            else:
                continue
        prob = 0.0
        if counter != 0:
            prob = counter/len(email_list)
        else:
            prob = 0.01  # TF=0.010.05spamham
        word_prob[word] = prob
    return word_prob


def filter(ham_word_pro, spam_word_pro, test_file):
    # (10)
    # right_ratespamham
    right = 0
    wrong = 0
    wrong_spam = 0
    test_paths = fileWalker(test_file)
    for test_path in test_paths:
        # 
        email_spam_prob = 0.0
        spam_prob = 0.5  # P(spam) = 0.5
        ham_prob = 0.5  # P(ham) = 0.5
        file_name = test_path.split('\\')[-1]
        prob_dict = {}
        words = set(email_parser(test_path))
        for word in words:  # wordP(spam|word)
            Psw = 0.0
            if word not in spam_word_pro:
                Psw = 0.4  # P(spam|new word) = 0.4 by Paul Graham
            else:
                Pws = spam_word_pro[word]  # P(word|spam)
                Pwh = ham_word_pro[word]  # P(word|ham)
                Psw = spam_prob*(Pws/(Pwh*ham_prob+Pws*spam_prob))
                # P(spam|word) = P(spam)*P(word|spam)/P(word)
                #              = P(spam)*P(word|spam)/(P(word|ham)*P(ham)+P(word|spam)*P(spam))
            prob_dict[word] = Psw
        numerator = 1
        denominator_h = 1
        for k, v in prob_dict.items():
            numerator *= v  # P1P2Pn = P(spam|word1)*P(spam|word2)**P(spam|wordn)
            denominator_h *= (1-v)  # (1-P1)(1-P2)(1-Pn) = (1-P(spam|word1))*(1-P(spam|word2))**(1-P(spam|wordn))
        email_spam_prob = round(numerator/(numerator+denominator_h), 4)
        # P(spam|word1word2wordn) = P1P2Pn/(P1P2Pn+(1-P1)(1-P2)(1-Pn))

        if email_spam_prob > 0.9:  # P(spam|word1word2wordn) > 0.9 spam
            print(file_name, 'spam', email_spam_prob)
            if file_name.split('_')[1] == '25.txt':
                print(prob_dict)
            if file_name.split('_')[0] == 'spam':  # 
                right += 1
            else:
                wrong += 1
                print('***********************Wrong Prediction***********************')
        else:
            print(file_name, 'ham', email_spam_prob)
            if file_name.split('_')[1] == '25.txt':
                print(prob_dict)
            if file_name.split('_')[0] == 'ham':  # 
                right += 1
            else:
                wrong += 1
                wrong_spam += 1  # spamham
                print('***********************Wrong Prediction***********************')

        # print(prob_dict)
    right_rate = right/(right+wrong)  # 
    if wrong != 0:
        wrong_spam_rate = [wrong_spam, wrong]  # [spamham]
    else:
        wrong_spam_rate = [-1]  # 0
    return right_rate, wrong_spam_rate


def main():
    # 
    right_rate_list = []
    wrong_spam_rate_list = []
    ham_file = r'D:\email\ham'
    spam_file = r'D:\email\spam'
    test_file = r'D:\email\test'
    for i in range(100):
        # 100
        test_set_select()  # 
        ham_list, ham_set = get_word(ham_file)
        spam_list, spam_set = get_word(spam_file)
        union_set = ham_set | spam_set  # 
        ham_word_pro = count_word_prob(ham_list, union_set)  # ham
        spam_word_pro = count_word_prob(spam_list, union_set)  # spam
        rig, wrg = filter(ham_word_pro, spam_word_pro, test_file)
        right_rate_list.append(rig)  # 
        wrong_spam_rate_list.append(wrg)  # spam->ham
        test_set_clear()  # 
    # 100
    x = range(100)
    y = right_rate_list
    plt.scatter(x, y)
    plt.title('Correct Rate of 100 Times')
    plt.show()
    # 100spam->ham
    print(wrong_spam_rate_list)
    return


if __name__ == '__main__':
    main()
```
